```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import csv

url = 'http://www.sec.gov/Archives/edgar/data/'
acc_no = input('Enter the document accession number :')
parts = acc_no.split('-')
print(parts)
print(type(parts[0]))
cik = parts[0]
year = parts[1]
no = parts[2]
CIK = cik.lstrip('0')
link = url+CIK+'/'+cik+year+acc_no+'/'+acc_no+'-index.html'
print(link)
request = requests.get(link)
html_doc = request.text
aw = 0
soup = BeautifulSoup(html_doc,'lxml')
a_tags = soup.find_all('a')

for a_tag in a_tags:
    if a_tag.get('href').endswith('_10q.htm'):
        print(a_tag.get('href'))
        link10q = ('https://www.sec.gov'+ a_tag.get('href'))
request1 = requests.get(link10q)        
html_doc1 = request1.text
soup1 = BeautifulSoup(html_doc1,'lxml')
table_tags = soup1.find_all('table')
print('########################')
print('*************************************')
print(len(table_tags))
calls_df = pd.read_html(html_doc1,header = 0)
print(type(calls_df))
print(len(calls_df))
for b in calls_df :
    #my_df = pd.DataFrame(calls_df)
    my_df = pd.DataFrame(b)
    aw = aw + 1
    file = cik+year+acc_no+str(aw)+'.csv'
    my_df.to_csv(file,index = False, header = False)








'''
csvfile ='/Users/rishabhjain/ADS_Assignment/csv1.csv'

with open(csvfile,'w') as output:
    writer = csv.writer(output,lineterminator='\n')
    for val in calls_df:
            print('Value    :',val)
            writer.writerow(val)
'''




'''
for table_tag in table_tags :
    print(table_tag)
    print('-----------')
'''
        
        
        
        
        
'''
---
request1 = requests.get(link10q)
print(request1)
html_doc1 = request1.text

soup = BeautifulSoup(html_doc1,'lxml')
print(soup.prettify())
---
''' 
```
